```python
import numpy as np
import matplotlib.pyplot as plt

# Soal 1 - membuat fungsi x untuk menyimpan persamaan
def f(x, expr):
    return eval(expr)

expr = input("Masukkan persamaan dalam (x): ")

# Soal 2 - membuat fungsi regula falsi
def regula_falsi(f, a, b, expr, tol=1e-5, max_iter=100):
    if f(a, expr) * f(b, expr) >= 0:
        print("Tidak ada akar di interval tersebut.")
        return None

    c = a # soal 3 - mendefinisikan nilai c = a
    for i in range(max_iter):
        # Soal 4 - mendifinisikan rumus regula falsi, menghitung nilai c
        c = (a * f(b, expr) - b * f(a, expr)) / (f(b, expr) - f(a, expr))
        # Soal 5 - mengecek apakah nilai c akurat
        if abs(f(c, expr)) < tol:
            break
        # Soal 6 - membuat interval baru
        if f(c, expr) * f(a, expr) < 0:
            b = c
        else:
            a = c
    return c

# Soal 7 - interval awal
a = 1
b = 2
# Soal 8 - mencari akar
akar = regula_falsi(f, a, b, expr)

# cek akar apakah ditemukan
if akar is not None:
  print(f"Akar yang ditemukan: {akar}")

  # Soal 9 - grafik fungsi akar
  x = np.linspace(0, 3, 400)
  func = lambda x: eval(expr)
  y = func(x)
  plt.plot(x, y, label=f'f(x) = {expr}')
  plt.axhline(0, color='black', lw=0.5, ls='--')
  plt.axvline(akar, color='red', lw=1, label=f'Akar: {akar:.5f}')
  plt.title('Grafik Fungsi dan Akar')
  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.legend()
  plt.grid()
  plt.show()
```
